```python
import matplotlib.pyplot as plt
```

```python
# how to save graph in matplotlib
val = [10,20,5,10,50,6,70]
plt.plot(val)
```

```python
plt.plot(val,marker='*',markersize=10,color='k',linestyle='--')
```

```python
plt.plot(val,marker='*',markersize=10,color='k',linestyle='--')
plt.xlabel('data')
plt.ylabel('count')
plt.title('simple graph')
plt.savefig('sample.png',dpi=300)
```

```python
# plot different color for line and marker
overs=range(1,11)
runs=[0,2,10,4,16,20,1,4,6,8]
```

```python
runs=np.cumsum(runs)
```

```python
runs
```

```python
plt.scatter(overs,runs,color='red')
```

```python
plt.plot(overs,runs,color='k') # to plot line only
```

```python
plt.plot(overs,runs,color='red',marker='*',markersize=10,linestyle='') # plot only marker put empty linestyle
```

```python
plt.plot(overs,runs,label='wickets')
plt.xlim([1,9])
plt.ylim([1,9])
plt.legend()
```
